# Preliminary Exploratory Data Analysis (EDA)
<hr>

## Reading dataset:

```r
heart_data <- read_csv("https://raw.githubusercontent.com/JackieHam/DSCI-100-project/main/report/data/heart.csv")
heart_data
```

## Cleaning and Wrangling:

We first change the column names to more understandable terms:

```r
new_column_names <- c("age", "sex", "chest_pain_type", "resting_blood_pressure", "serum_cholesterol", "fasting_blood_pressure>120mg/dl", "resting_electrocardiographic", "max_heart_rate", "exercies_induced_angina", "oldpeak", "slope_of_peak_exercise_ST", "major_vessels_num", "thal", "heart_disease")
colnames(heart_data) <- new_column_names
heart_data<- heart_data |>
            mutate(target = as_factor(heart_disease))
```

```r
heart_data
```

The column names are now more intuitive for people who are unfamiliar with the dataset.

## Method

Since the target class, whether the patient has heart disease, is a binary class, we decide to use KNN classifier.

# Feature Selection
<hr>

## 1. Number of features

Since we are planning to use KNN classifier to build our model, we should not have too many features.

This is because increasing the number of features will increase the dimension of the problem. Since KNN classifier makes prediction based on Euclidian distance, it will perform pooly in high dimensions (curse of dimensionality).

Therefore, we will only choose 2 features to build our model and eventually make predictions about whether the patient is likely to have heart disease on unseen data.

## 2. Dropping Binary and Discrete Features

Since KNN is a distance-based algorithm and assumes that all features contribute equally to the distance calculation, we decided to drop the binary features and discrete features. This is because and discrete features might not adhere to this assumption, especially if some features are more informative than others.

Therefore, the following features are dropped:

- sex
- chest_pain_type
- fasting_blood_pressure>120mg/dl
- resting_electrocardiographic
- exercies_induced_angina
- slope_of_peak_exercise_ST
- major_vessels_num
- thal

## 3. Splitting of dataset

We split our heart_data_scaled into training and testing data, with training data taking up 75% of the total data available:

The 75% trainig data proportion helps ensure that we have sufficient training data to build an accurate predictive model. 

At the same time, we have 25% of the testing data which enables us to evaluate the performance of our model based on test score. If test score turns out to be poor, we will make necessary adjustments at the training stage to improve performance of our model as much as possible.

```r
set.seed(1234)
hd_split <- initial_split(heart_data, prop = 0.75, strata = target)
hd_training <- training(hd_split)
hd_testing <- testing(hd_split)
```

## 4. Feature Selection Among the Remaing Features

Features remaining are:

- age
- resting_blood_pressure
- serum_cholesterol
- max_heart_rate
- oldpeak

To make sure that all continuous features contribute equally to the distance equally, we create a recipe that normalizes the data:

```r
heart_data_recipe <- recipe(target ~ age + resting_blood_pressure + serum_cholesterol + max_heart_rate + oldpeak, data = hd_training) |>
                       step_scale(all_predictors()) |>
                       step_center(all_predictors())
```

Then, we pass our heart_data into the recipe, to create a new dataframe that contains only scaled continuous feature:

```r
hd_training_scaled <- heart_data_recipe |>  
                           prep() |> 
                           bake(hd_training)
```

```r
hd_training_scaled
```

### Final Decision for Features

After looking at the above graphical representations, we've narrowed down to two following pairs of features and will choose one of the two:
- max_heart_rate and serum_cholesterol
- max_heart_rate and oldpeak

Eventually, we decided to choose max_heart_rate and serum_cholesterol. The main reason is that it seems that observation data is not very evenly distributed for oldpeak. Looking at the graph, there are way more observations with oldpeak value = -1 than that with other oldpeak values. We feel that this may have a negative effect on the accuracy of predictions in our model trained using the oldpeak feature.

# KNN Model Building
<hr>

* Features we'll use:
    - serum_cholesterol
    - maximum heart rate achieved
    - (target: heart_disease)

## Hyperparameter Tuning

We need to tune the hyperparameter (neighbors). We carry out the tuning process:

```r
knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
       set_engine("kknn") |>
       set_mode("classification")
```

```r
set.seed(1234)
hd_vfold <- vfold_cv(hd_training, v = 5, strata = heart_disease)

knn_results <- workflow() |>
       add_recipe(heart_data_recipe) |>
       add_model(knn_tune) |>
       tune_grid(resamples = hd_vfold, grid = 10) |>
       collect_metrics()
```

```r
options(repr.plot.width = 8, repr.plot.height = 6)
accuracies <- knn_results |> 
       filter(.metric == "accuracy")

accuracy_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       labs(x = "Neighbors", y = "Accuracy Estimate") +
       scale_x_continuous(breaks = seq(0, 14, by = 1)) + 
       scale_y_continuous(limits = c(0.4, 1.0)) 
accuracy_versus_k
```

Based on the diagram, it looks like either neighbors = 1 or neighbors = 2 give the highest accuracy estimate. Since even number can lead to a draw, we choose neighbors = 1.

## Model Building

Using neighbors = 1, we build our knn_spec model:

```r
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 1) |>
  set_engine("kknn") |>
  set_mode("classification")
```

## Fitting Training Data

```r
knn_fit <- workflow() |>
  add_recipe(heart_data_recipe) |>
  add_model(knn_spec) |>
  fit(data = hd_training)
```

## Prediction Making on Test Data

```r
hd_test_predictions <- predict(knn_fit, hd_testing) |>
  bind_cols(hd_testing)
hd_test_predictions
```

## Obtaining Accuracy Scores of Predictions

```r
hd_test_predictions |>
  metrics(truth = target, estimate = .pred_class) |>
  filter(.metric == "accuracy")
```
